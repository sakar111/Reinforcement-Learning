```python
import numpy as np
import random
```

A Simple Bandit Algorithm Page-32

initialize epsilon(e)

Initialize, for a = 1 to k:
  Q(a) = 0
  N(a) = 0
Loop for i=1 to N:
  

```python
def bandit_algorithm(e):
  Q = np.zeros(k) # Action Value for every action
  N = np.zeros(k) # Action counter

  Average_Reward = 0

  Average_Reward_Data = []

  for step in range(total_steps):

    if random.random() < e or step == 0:
      index = random.randint(0,9)
    else:
      index = np.argmax(Q)

    Reward = np.random.normal(true_actionValue[index], np.sqrt(actionVariance[index]))

    N[index] += 1
    Q[index] = Q[index] + 1 / N[index] * (Reward - Q[index])

    Average_Reward = Average_Reward + 1/(step + 1) * (Reward - Average_Reward)

    Average_Reward_Data.append(Average_Reward)

  return Average_Reward_Data
```

k = no.of armed bandits

q*(a) is the action value of each bandit for a = 1..k which is selected with gaussian distribution of mean u and variance v

```python
k = 10
u = 0
v = 1
true_actionValue = np.random.normal(u, np.sqrt(v), k)
actionVariance = np.ones(k)

e0 = 0
e1 = 0.1
e2 = 0.2
e3 = 0.3
e4 = 0.05

total_steps = 1000

Average_Reward_Data_e0 = bandit_algorithm(e0)
Average_Reward_Data_e1 = bandit_algorithm(e1)
Average_Reward_Data_e2 = bandit_algorithm(e2)
Average_Reward_Data_e3 = bandit_algorithm(e3)
Average_Reward_Data_e4 = bandit_algorithm(e4)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the Average_Reward over steps
plt.plot(np.arange(total_steps), Average_Reward_Data_e0, label = 'e = 0.0')
plt.plot(np.arange(total_steps), Average_Reward_Data_e1, label = 'e = 0.1')
plt.plot(np.arange(total_steps), Average_Reward_Data_e2, label = 'e = 0.2')
plt.plot(np.arange(total_steps), Average_Reward_Data_e3, label = 'e = 0.3')
plt.plot(np.arange(total_steps), Average_Reward_Data_e4, label = 'e = 0.05')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
plt.title('Average Reward over Steps')
plt.grid(True)

plt.legend()

plt.show()
```
